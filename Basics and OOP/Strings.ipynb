```python
#Frequency counter
text=input('Give text')
term=input('What you want to search for')

counter=0
for i in text:
    if i==term:
        counter += 1
    
print(counter)
```

```python
#Remove text step by step from string by ignoring given 
s=input('Give text')
term=input('What you want to remove for')
result=''

for i in s:
    if i!=term:
        result=result+i

print(result)
```
