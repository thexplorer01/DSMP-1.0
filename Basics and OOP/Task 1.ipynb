### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Changed int() to float() and added spaces/colons to the prompts
p1x = float(input('Enter x co-ordinate of 1st point: '))
p2x = float(input('Enter x co-ordinate of 2nd point: '))
p1y = float(input('Enter y co-ordinate of 1st point: '))
p2y = float(input('Enter y co-ordinate of 2nd point: '))

# Your math is perfect!
euclidean = ((p2x - p1x)**2 + (p2y - p1y)**2)**0.5

print('The distance is:', euclidean)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
first_term=int(input('Enter first term'))
second_term=int(input('Enter second term'))
n=int(input('value of n'))
d=second_term-first_term

an=(first_term+(n-1)*d)
print(an) 
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
import math
h_t= float(input('Enter height of the tank'))
l_t= float(input('Enter length of the tank'))
b_t= float(input('Enter breadth of the tank'))

h_g= float(input('Enter height of the glass'))
r_g= float(input('Enter radius of the glass'))

vol_tank=(h_t*l_t*b_t)
vol_glass=(3.14*r_g*r_g*h_g)

glasses=vol_tank/vol_glass
print('The number of glass milk obtained are',math.floor(glasses))
````
