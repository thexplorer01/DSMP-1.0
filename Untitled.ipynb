```python
#Guessing game while is for infinite spawn or not known
import random 
jackpot = random.randint(1,100)

guess = int(input('Try your luck'))
counter =1

while guess!=jackpot:
    if guess<jackpot:
        print('guess higher')
    else: print('guess lower')
    guess = int(input('Try your luck'))
    counter+=1
else: 
  print('sahi jawab')
print('you took',counter,'attempts')
```

```python
# Star pattern
rows = int(input('Enter no: '))

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print('*', end='')
    print()  # Moved back one indentation level
```

```python

rows = int(input('Enter no of rows'))
for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j,end='')
    for k in range(i-1,0,-1):
        print(k,end='')

    print()

                            

      
    
```

```python
#Loop for prime number with break statement to prevent further non eligible numbers
lower=int(input('Enter lower limit number'))
upper=int(input('Enter upper limit number'))
for i in range(lower,upper+1):
    for j in range(2,i):
        if i%j==0:
            break
        else:
            print(i)
    
```
